## 1.2. List Vs Numpy Array 

## -  List
       1. Different datatypes.
       2. Operations execute not in the vector form it need loop. 
       3. Python built-in package
       4. Occupies more memory and slow in computation.
       5. Not so convenient to use for higher order matehmatical operations.
       
## - Numpy Array
       1. Similar datatypes.
       2. Operations can execute also in the vector form.
       3. Install Numpy package separately after installation of Python.( command prompt > pip/pip3 install numpy)   
       4. Occupies less memory and fast in computation.
       5. More convenient to use for higher order matehmatical operations.

## 2.6. Image Geometric transformations

```python
# Image Geometrical Operations # Traslate
import numpy as np
import cv2

image=cv2.imread('C:/Users/pchak/Desktop/OpenCV class stuff/jeep.jpg')

def translate(image,x,y): # shifting
    trans_matrix=np.float32([[1,0,x],[0,1,y]])
    shifted=cv2.warpAffine(image,trans_matrix,(image.shape[1],image.shape[0]))
    return shifted;

#Image Translate
trans1=translate(image,0,100)    
trans2=translate(image,0,-100)
trans3=translate(image,100,0)
trans4=translate(image,-100,0)

cv2.imshow("original image",image)
cv2.waitKey(0)
cv2.imshow("Down shift image",trans1)
cv2.waitKey(0)
cv2.imshow("Top shift image",trans2)
cv2.waitKey(0)
cv2.imshow("right shift image",trans3)
cv2.waitKey(0)
cv2.imshow("left shift image",trans4)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# Image Geometrical Operations # resize
import numpy as np
import cv2


image=cv2.imread('C:/Users/pchak/Desktop/OpenCV class stuff/jeep.jpg')


def resize(image,width=None,height=None):
    (h, w) = image.shape[:2]
    if width is None and height is None:
        return image
    if width is None:
        r = height / float(h)
        dim = (int(w * r), height)
    else:
        r = width / float(w)
        dim = (width, int(h * r))

    resized=cv2.resize(image,dim,interpolation=cv2.INTER_AREA)
    # interpolation method, which is the algorithm working behind the scenes to handle how the actual image is resized.
    #   cv2.INTER_LINEAR, cv2.INTER_CUBIC, and cv2.INTER_NEAREST.
    return resized;

# Image resize 
resizeimg1=resize(image,width=250,height=250)
resizeimg2=resize(image,width=800,height=800)
cv2.imshow("Original Image",image)
cv2.waitKey(0)
cv2.imshow("Resized Image1:",resizeimg1)
cv2.waitKey(0)
cv2.imshow("Resized Image2:",resizeimg2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Image Geometrical Operations rotation
import numpy as np
import cv2
import time

image=cv2.imread('C:/Users/pchak/Desktop/OpenCV class stuff/jeep.jpg')

def rotate(image,angle,scale):
    (w,h)=image.shape[:2]
    
    rotate_m=cv2.getRotationMatrix2D((w//2,h//2),angle,scale) # (ratation point,angle,
    rotated=cv2.warpAffine(image,rotate_m,(w,h))
    
    return rotated;

# Image Rotation
j=0;
for i in range(0,405,45):
    rotate1=rotate(image,i,1.0)
    cv2.imshow("Rotated images",rotate1)
    cv2.waitKey(0)
    if j==8:
        break;
    j+=1;

cv2.destroyAllWindows()
```
